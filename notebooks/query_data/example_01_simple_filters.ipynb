# Generate Random Data

```python
import json
with open('/workspaces/work_utils2/notebooks/creating_random_data/example_03.json', 'r') as f:
    specs = json.load(f)
df = rd.create_dataframe(specs, n_rows=10000, random_seed=42)
df.sample(5)
```

# Demonstrate Query Tools

## Basic Filters

```python
active_filter = {
    "type": "equals",
    "column": "active",
    "value": True
}

active_mask = qt.filters.apply_filter(df, active_filter)
df[active_mask].sample(5)
```

```python
category_filter = {
    "type": "in",
    "column": "product_category",
    "values": ["Electronics", "Books"]
}

category_mask = qt.filters.apply_filter(df, category_filter)
df[category_mask].sample(5)
```

```python
score_filter = {
    "type": "between",
    "column": "score",
    "min_value": 70,
    "max_value": 90
}

score_mask = qt.filters.apply_filter(df, score_filter)
df[category_mask].sample(5)
```

```python
combined_mask = active_mask & category_mask & score_mask
df[combined_mask].sample(5)
```

## Date Filters

```python
q1_filter = {
    "type": "between",
    "column": "open_date",
    "min_value": "2024-01-01",
    "max_value": "2024-03-31"
}

# Apply the filter and show results
q1_mask = qt.filters.apply_filter(df, q1_filter)
df[q1_mask].sample(5)
```

```python
# Filter 2: Month filter
june_filter = {
    "type": "date_month",
    "column": "open_date",
    "month": 6
}

# Apply the filter and show results
june_mask = qt.filters.apply_filter(df, june_filter)
df[june_mask].sample(5)
```

```python
# Filter 3: Weekday filter (Monday = 0)
monday_filter = {
    "type": "date_weekday",
    "column": "open_date",
    "weekday": 0
}

# Apply the filter and show results
monday_mask = qt.filters.apply_filter(df, monday_filter)
df[monday_mask].sample(5)
```

## Complex Filter

```python
complex_filter = {
    "type": "and",
    "filters": [
        # Must be electronics
        {
            "type": "equals",
            "column": "product_category",
            "value": "Electronics"
        },
        # Must be high value (> 150)
        {
            "type": "greater_than",
            "column": "price",
            "value": 150
        },
        # Must be either discounted OR have high score
        {
            "type": "or",
            "filters": [
                {
                    "type": "equals",
                    "column": "is_discounted",
                    "value": True
                },
                {
                    "type": "greater_than",
                    "column": "score",
                    "value": 80
                }
            ]
        }
    ]
}

# Apply the filter and show results
complex_mask = qt.filters.apply_filter(df, complex_filter)
df[complex_mask].sample(5)
```
